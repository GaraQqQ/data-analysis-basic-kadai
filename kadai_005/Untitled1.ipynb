```python
import numpy as np
import matplotlib.pyplot as plt

# データ生成
x = np.linspace(0, 10, 100)
y_line = x * 2 + 1                # 折れ線グラフ用
y_sin = np.sin(x)                 # sin関数用
data_hist = np.random.randn(1000) # ヒストグラム用
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

# 描画キャンバスを2x2分割
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 折れ線グラフ
axs[0, 0].plot(x, y_line, color='blue')
axs[0, 0].set_title('折れ線グラフ')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

# sin関数
axs[0, 1].plot(x, y_sin, color='green')
axs[0, 1].set_title('sin関数')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('sin(x)')

# ヒストグラム
axs[1, 0].hist(data_hist, bins=30, color='orange')
axs[1, 0].set_title('ヒストグラム')
axs[1, 0].set_xlabel('値')
axs[1, 0].set_ylabel('頻度')

# 散布図
axs[1, 1].scatter(x_scatter, y_scatter, color='red')
axs[1, 1].set_title('散布図')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

plt.tight_layout()
plt.show()
```
